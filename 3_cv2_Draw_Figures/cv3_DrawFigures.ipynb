```python
import numpy as np
import cv2

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw a diagonal blue line with thickness of 5 px
img = cv2.line(img,(0,0),(511,511),(255,0,0),15)
```

```python
from matplotlib import pyplot as plt

plt.imshow(img)
plt.show()
```

```python
img = cv2.rectangle(img,(384,10),(510,128),(0,255,0),8)
plt.imshow(img)
plt.show()
```

```python
img = cv2.circle(img,(447,70), 63, (0,0,255), -1)
plt.imshow(img)
plt.show()
```

```python
img = np.zeros((512,512,3), np.uint8)
img = cv2.ellipse(img,(256,256),(100,50),45,0,270,255,20)
plt.imshow(img)
plt.show()
```

```python
img = np.zeros((512,512,3), np.uint8)
pts = np.array([[10,5],[200,300],[470,200],[500,10]], np.int32)
pts = pts.reshape((-1,1,2))
img = cv2.polylines(img,[pts],True,(0,255,255))
plt.imshow(img)
plt.show()
```

```python
img = np.zeros((512,512,3), np.uint8)
font = cv2.FONT_ITALIC
cv2.putText(img,'OpenCV',(10,470), font, 4,(255,255,255),10,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```
