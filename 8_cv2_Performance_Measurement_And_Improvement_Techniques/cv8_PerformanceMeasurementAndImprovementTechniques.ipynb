```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

e1 = cv2.getTickCount()

# your code execution
img = cv2.imread('hills.jpg') #0.008014962
plt.imshow(img) #0.019024376

e2 = cv2.getTickCount() # return the number of clock-cycles after a reference event 
time = (e2 - e1)/ cv2.getTickFrequency() # returns the frequency of clock-cycles, or the number of clock-cycles per second.
print(time)
```

```python
img1 = cv2.imread('hills.jpg')

e1 = cv2.getTickCount()
for i in range(7,42,10):
    img1 = cv2.medianBlur(img1,i)
plt.imshow(img1)
e2 = cv2.getTickCount()
t = (e2 - e1)/cv2.getTickFrequency()
print (t)
```

- Avoid using loops in Python as far as possible, especially double/triple loops etc. They are inherently slow.

- Vectorize the algorithm/code to the maximum possible extent because Numpy and OpenCV are optimized for vector operations.

- Exploit the cache coherence.

- Never make copies of array unless it is needed. Try to use views instead. Array copying is a costly operation.
