```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('rakushka.jpg',0)
rows = img.shape[0]
cols = img.shape[1]

M = np.float32([[1,0,-100],[0,1,100]])
dst = cv2.warpAffine(img,M,(cols,rows))

plt.imshow(dst)
```

```python
#set the center of rotation to position (1,1)
M = cv2.getRotationMatrix2D((1,1),45,1)
#if you want to set the center in teh center then set (cols/2,rows/2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

# Affine Transformation

```python
# rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

# Perspective Transformation

For perspective transformation, you need a 3x3 transformation matrix. Straight lines will remain straight even after the transformation. To find this transformation matrix, you need 4 points on the input image and corresponding points on the output image. Among these 4 points, 3 of them should not be collinear. 

```python
pts1 = np.float32([[56,65],[468,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))
plt.imshow(dst)
```
