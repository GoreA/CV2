# 2D Convolution ( Image Filtering )

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('red_blue.jpg')
```

```python
kernel = np.ones((15,15),np.float32)/25
dst = cv2.cvtColor(cv2.filter2D(img, -1, kernel), cv2.COLOR_BGR2RGB)
plt.imshow(dst)
```

# Image Blurring (Image Smoothing)

```python
blur = cv2.cvtColor(cv2.blur(img,(15,15)), cv2.COLOR_BGR2RGB)
plt.imshow(blur)
```

```python
blur = cv2.cvtColor(cv2.GaussianBlur(img,(15,15),0), cv2.COLOR_BGR2RGB)
plt.imshow(blur)
```

```python
blur = cv2.cvtColor(cv2.medianBlur(img,15), cv2.COLOR_BGR2RGB)
plt.imshow(blur)
```
