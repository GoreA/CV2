```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('silhouette.jpg')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray, 127, 255,0)
_, contours, hierarchy = cv2.findContours(thresh,2,1)
cnt = contours[0]

hull = cv2.convexHull(cnt,returnPoints = False)
defects = cv2.convexityDefects(cnt,hull)

for i in range(defects.shape[0]):
    s,e,f,d = defects[i,0]
    start = tuple(cnt[s][0])
    end = tuple(cnt[e][0])
    far = tuple(cnt[f][0])
    cv2.line(img,start,end,[0,255,0],2)
    cv2.circle(img,far,5,[0,0,255],-1)

plt.figure(figsize=(12, 10))
plt.imshow(img)
```

This function finds the shortest distance between a point in the image and a contour. It returns the distance which is negative when point is outside the contour, positive when point is inside and zero if point is on the contour.

```python
dist = cv2.pointPolygonTest(cnt, (50, 50), True)
print(dist)
```

Enables us to compare two shapes, or two contours and returns a metric showing the similarity. 

```python
img = cv2.imread('circles.jpg')
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 127, 255,0)
_, contours, hierarchy = cv2.findContours(thresh,2,1)
cnt0 = contours[0]
cnt1 = contours[1]

ret = cv2.matchShapes(cnt1, cnt0, 1, 0)
print(ret)
ret = cv2.matchShapes(cnt0, cnt1, 1, 0)
print(ret)
```
